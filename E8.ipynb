```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("marks.csv")

print(df.head())
print(df.tail())

print("describe() gives summary statistics like count, mean, std, min, max, and quartiles.")
print(df.describe())

print(df.iloc[2:6])
print(df[df['M3'] > 84])
print(df[df.isnull().any(axis=1)])
print(df.dropna())
print(df.sort_values(by='M1', ascending=False))

df.set_index('NAME')[['M1', 'M2', 'M3']].plot(kind='bar', figsize=(10, 6))
plt.title("Student Marks in M1, M2, M3")
plt.ylabel("Marks")
plt.xlabel("Name")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print(df.iloc[3:7][['M2', 'M3']])
```
